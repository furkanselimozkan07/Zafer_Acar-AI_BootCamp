```python
## Bar Chart
# Use the following data to make your plot:

from numpy import arange, array
x = arange(1, 13)
y = array([7, 8, 5, 5, 3, 4, 4, 6, 9, 7, 8, 8])
```

1. Add a bar chart to your plot, using `x` and `y` above. Fix it so there is no outline (edgecolor), set the color of the bars to green, and set their width to 0.25.
2. Fix your X-axis to go from 1 to 13.
3. Fix your Y-axis to go from 0 to 10.
4. Add a label to your X-axis that says "month of the year".
5. Add a label to your Y-axis that says "letters in month name".
6. Add a title to your plot, "Leters in the Name of Each Month".
7. Display your plot.

```python
## Bar Chart
# Use the following data to make your plot:

from numpy import arange, array
import matplotlib.pyplot as plt

# Verilen x ve y verileri
x = arange(1, 13)
y = array([7, 8, 5, 5, 3, 4, 4, 6, 9, 7, 8, 8])


# --- Grafik Oluşturma ---
# plt.bar() fonksiyonu ile bir bar grafiği oluşturuyoruz.
# İstenen özellikleri parametre olarak giriyoruz:
# color='green': Barların rengini yeşil yapar.
# width=0.25: Barların genişliğini 0.25 olarak ayarlar.
# edgecolor='none': Barların kenar çizgisini (outline) kaldırır.
plt.bar(x, y, color='green', width=0.25, edgecolor='none')


# --- Eksen ve Başlık Ayarları ---
# X ekseninin sınırlarını 1'den 13'e ayarlar.
plt.xlim(1, 13)
# Y ekseninin sınırlarını 0'dan 10'a ayarlar.
plt.ylim(0, 10)

# X eksenine "month of the year" etiketini ekler.
plt.xlabel("month of the year")
# Y eksenine "letters in month name" etiketini ekler.
plt.ylabel("letters in month name")

# Grafiğe "Letters in the Name of Each Month" başlığını ekler.
plt.title("Letters in the Name of Each Month")


# --- Grafiği Gösterme ---
# Hazırlanan grafiği ekranda gösterir.
plt.show()

```

```python
## Scatter and Line Plots
# Use the following data:

from numpy import arange, cos, pi
x = arange(0.0, 2 * pi, pi / 16.0)
y = 2 * cos(x)
```

1. Add a line plot, using `x` and `y` above. Set the width of the line to 3, set the color to green, set the style to dashed, with a label of "approx".
2. Add a scatter plot, using `x` and `y` above. Set the size to 1000, set the edgecolor to none, set the color to blue, and set the label to "discrete".
3. Add an X label of "x".
4. Add a Y label of "Y".
5. Add a title of "2 * cos(x)".
6. Save your plot to a file named "cos.png".

```python
# Solution
## Scatter and Line Plots
# Use the following data:

from numpy import arange, cos, pi
import matplotlib.pyplot as plt

# X ve Y eksenleri için veri tanımlaması
# x, 0'dan 2*pi'ye kadar belirli adımlarla artan bir dizidir.
x = arange(0.0, 2 * pi, pi / 16.0)
# y, x'in her bir değeri için 2 * cos(x) fonksiyonunun sonucudur.
y = 2 * cos(x)


# --- Grafik Oluşturma ---
# Çizgi grafiği (Line Plot) ekleniyor.
# linewidth=3: Çizgi kalınlığını 3 olarak ayarlar.
# color='green': Rengini yeşil yapar.
# linestyle='dashed': Çizgi stilini kesikli yapar.
# label='approx': Grafiğin etiketini belirler (lejant için).
plt.plot(x, y, linewidth=3, color='green', linestyle='dashed', label='approx')

# Dağılım grafiği (Scatter Plot) ekleniyor.
# s=1000: Noktaların boyutunu 1000 olarak ayarlar.
# edgecolor='none': Noktaların kenar rengini kaldırır.
# color='blue': Noktaların rengini mavi yapar.
# label='discrete': Grafiğin etiketini belirler (lejant için).
plt.scatter(x, y, s=1000, edgecolor='none', color='blue', label='discrete')


# --- Eksen ve Başlık Ayarları ---
# X eksenine "x" etiketini ekler.
plt.xlabel("x")
# Y eksenine "Y" etiketini ekler.
plt.ylabel("Y")

# Grafiğe "2 * cos(x)" başlığını ekler.
plt.title("2 * cos(x)")

# Grafikteki etiketleri göstermek için bir lejant ekler.
plt.legend()


# --- Grafiği Kaydetme ve Gösterme ---
# Oluşturulan grafiği "cos.png" adıyla bir dosyaya kaydeder.
plt.savefig('cos.png')

# Hazırlanan grafiği ekranda gösterir.
plt.show()
```

```python
## Multiplot, Histogram, and Errorbar
# Use the following data:

from numpy import random
mu = 10500
sigma = 250
values = mu + sigma * random.randn(10000)
x = mu + sigma * random.randn(100)
y = mu + sigma * random.randn(100)
```

1. Use `subplot` to initialize the first plot (left column) in a two-column subplot image.
2. Create a histogram of `values` with 25 columns, set the facecolor to green, and set the edgecolor to none.
3. Give the plot a Y-label of "count" of values".
4. Use `subplot` to intiialize the second plot (right column) in a two-column subplot image.
5. Create a scatter plot of `x` and `y`, set the area to 300, set alpha to 0.25, set edgecolor to none, set color equal to `random.rand(100)`.
6. Give the plot a X-label of "x".
7. Give the plot a Y-label of "y".
8. Set the x-limits of the plot to: 10,000 to 11,000.
9. Set the y-limits of the plot to: 10,000 to 11,000.
10. Add a grid to the plot.
11. Show the plot.

```python
# Solution
## Multiplot, Histogram, and Errorbar
# Use the following data:

from numpy import random
import matplotlib.pyplot as plt

# Veri tanımlaması
mu = 10500
sigma = 250
values = mu + sigma * random.randn(10000)
x = mu + sigma * random.randn(100)
y = mu + sigma * random.randn(100)


# --- Çoklu Grafik Alanını Ayarlama ---
# 1 satır ve 2 sütundan oluşan bir grafik alanı oluşturuyoruz.

# --- Sol Sütun: Histogram ---
# subplot(1, 2, 1) -> 1 satır, 2 sütunluk alanın 1. grafiğini seçer.
plt.subplot(1, 2, 1)

# 'values' verisiyle 25 sütunlu bir histogram oluşturuluyor.
# facecolor='green': Sütunların rengini yeşil yapar.
# edgecolor='none': Sütunların kenar çizgisini kaldırır.
plt.hist(values, bins=25, facecolor='green', edgecolor='none')
# Y eksenine etiket ekleniyor.
plt.ylabel("count of values")


# --- Sağ Sütun: Scatter Plot ---
# subplot(1, 2, 2) -> 1 satır, 2 sütunluk alanın 2. grafiğini seçer.
plt.subplot(1, 2, 2)

# x ve y verileriyle bir dağılım grafiği oluşturuluyor.
# s=300: Noktaların alanını 300 olarak ayarlar.
# alpha=0.25: Noktaların şeffaflığını %25 olarak ayarlar.
# edgecolor='none': Noktaların kenar çizgisini kaldırır.
# c=random.rand(100): Her bir noktaya rastgele bir renk atar.
plt.scatter(x, y, s=300, alpha=0.25, edgecolor='none', c=random.rand(100))

# Eksen etiketleri ekleniyor.
plt.xlabel("x")
plt.ylabel("y")

# Eksen limitleri ayarlanıyor.
plt.xlim(10000, 11000)
plt.ylim(10000, 11000)

# Grafiğe bir grid (ızgara) ekleniyor.
plt.grid(True)


# --- Grafiği Gösterme ---
# Hazırlanan çoklu grafiği ekranda gösterir.
plt.show()
```
